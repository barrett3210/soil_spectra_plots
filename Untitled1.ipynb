```python
import pandas as pd
import numpy as np
import h5py
import datetime
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

file_name = "data/210407/soc_210407.h5"
h5_file = h5py.File(file_name, "r")
```

```python
insertions = list(h5_file['session001/cal001'].keys())
vis_waves = h5_file['session001/cal001'].attrs['spec1_wavelengths_vector'][:]
ftir_waves = h5_file['session001/cal001'].attrs['spec2_wavelengths_vector'][:]
```

```python
ins1 = h5_file['session001/cal001/ins001/derived/absorbance_depth'][:, :512]
```

```python
ins1.shape
```

```python
vis_waves.shape
```

```python
depth = h5_file['session001/cal001/ins001/derived/absorbance_depth'][:, -1]
```

```python
y = depth
x = vis_waves
z = savgol_filter(ins1, 15, 3, axis=1)
X, Y = np.meshgrid(x, y)
```

```python
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection='3d')
ax.set_ylim(60, 0)
# ax.view_init(40, 275)
# ax.contour3D(X, Y, z, 100)
ax.plot_surface(X, Y, z, cmap='viridis', edgecolor='none')
```

```python
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection='3d')
ax.set_ylim(60, 0)

for i in range(z.shape[0]):
    depth_fraction = (abs(depth[i]/(depth.max() - depth.min())) * 0.5)
    color = (0.7, depth_fraction, 0.2)
    ax.plot(vis_waves, z[i, :], zs=depth[i], zdir='y', color=color, lw=7.5, alpha=0.7)
```

```python
for insertion in insertions:
    ins1 = h5_file[f'session001/cal001/{insertion}/derived/absorbance_depth'][:, :512]
    depth = h5_file[f'session001/cal001/{insertion}/derived/absorbance_depth'][:, -1]
    y = depth
    x = vis_waves
    z = savgol_filter(ins1, 15, 3, axis=1)
    X, Y = np.meshgrid(x, y)
    fig = plt.figure(figsize=(15,15))
    ax = plt.axes(projection='3d')
    ax.set_ylim(90, 0)
    # ax.view_init(40, 275)
    # ax.contour3D(X, Y, z, 100)
    ax.plot_surface(X, Y, z, cmap='viridis', edgecolor='none')
    
```
