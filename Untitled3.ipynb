```python
def plot_spectrum_mean_std(ax, waves, spectra):
    ax.plot(waves, spectra.mean(axis=0), 'tab:blue', lw=4)
    ax.plot(waves, spectra.mean(axis=0) - spectra.std(axis=0), 'tab:blue', alpha=0.5)
    ax.plot(waves, spectra.mean(axis=0) + spectra.std(axis=0), 'tab:blue', alpha=0.5)
```
